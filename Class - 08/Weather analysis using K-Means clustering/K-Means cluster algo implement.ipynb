```python
df.rename(columns={'Temperature (°C)': 'temp', 'Humidity (%)': 'hum', 'Wind Speed (km/h)':'ws', 'Precipitation (mm)':'prect', 'Pressure (hPa)': 'presure', 'Cloud Cover (%)':'cloud'}, inplace=True)
```

```python
df
```

# Scaling dataset

```python
sclare = StandardScaler()
```

```python
df['temp'] = sclare.fit_transform(df[['temp']])
df['hum'] = sclare.fit_transform(df[['hum']])
df['ws']  = sclare.fit_transform(df[['ws']]) 
df['prect']  = sclare.fit_transform(df[['prect']])  
df['presure']  = sclare.fit_transform(df[['presure']])   
df['cloud']  = sclare.fit_transform(df[['cloud']])     
```

```python
df
```

```python
sns.pairplot(df[['temp', 'hum', 'ws', 'prect']])
```

```python
plt.scatter(df.temp, df.hum, color='red', marker='o', label="temparature and humidity")

plt.xlabel('temprature')
plt.ylabel('humidity')
plt.title('Temprature and Humidity visializatioin')
plt.legend()
```
